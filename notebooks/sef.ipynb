```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
vrel = 220 / 300000

def sef(alpha):
	return 2 * np.pi * alpha / vrel / (1 - np.exp(-2 * np.pi * alpha / vrel))
```

```python
alpha_axis = np.logspace(-3, -1)
plt.title("Sommerfeld enhancement factor with $v_{rel}=220$km/s")
plt.xlabel("$\\alpha$")
plt.ylabel("S")
plt.xscale("log")
plt.grid(which="both")
plt.plot(alpha_axis, sef(alpha_axis))
```
